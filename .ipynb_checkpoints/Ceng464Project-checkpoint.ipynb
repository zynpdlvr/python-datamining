```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_excel('Course Project - Data for Classification - Electric Vehicles.xls')
df.head()
```

```python
print(df['Q16'].unique()) 
```

```python
df['Q16'].value_counts() 
```

```python
df.replace('?', np.nan, inplace=True)

# Q16 sütununu one-hot encoding ile dönüştür
df = pd.get_dummies(df, columns=['Q16'], prefix=['Q16'])

# Dönüştürülmüş veriyi yazdır
print(df)
```

```python
df.head()
```

```python
from sklearn.impute import KNNImputer
import numpy as np



# Kullanmak istediğiniz imputasyon sütunlarını seçin
imputation_columns = df.columns

# Denenecek k değerleri
k_values = [1, 3, 5, 7, 9]

# En iyi k değerini ve performansını saklamak için değişkenler
best_k = None
best_score = 0

# Her bir k değeri için KNN imputasyonu yapın
for k in k_values:
    imputer = KNNImputer(n_neighbors=k)
    df_imputed = imputer.fit_transform(df)

    # Eksik değerleri doldurulmuş veriyi kullanarak bir sınıflandırma modeli eğitme
    # Bu kısmı atlıyoruz, sadece eksik değerleri dolduruyoruz.

    # (Opsiyonel) Modelin performansını değerlendirme
    # Bu kısmı atlıyoruz, sadece eksik değerleri dolduruyoruz.

    # En iyi k değerini sakla
    best_k = k

# En iyi k değeri ile tüm veri setini doldurun
best_imputer = KNNImputer(n_neighbors=best_k)
df_best_imputed = pd.DataFrame(best_imputer.fit_transform(df), columns=df.columns)

# Eksik değerleri doldurulmuş veriyi yazdır
print(df_best_imputed)
print(best_k)
```
